# Read, Resize, Flip and Save Images Using OpenCV

**Features:**
- Read an image using OpenCV.
- Convert the OpenCV BGR image to RGB image.
- Display the image using Matplotlib.
- Resize the image by passing height and width values.
- Change the image size to width ratio & height ratio.
- Perform different flips on the image.
- Save the image to the local drive.
- Display the image in a seperate window.

```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

## Read an image using OpenCV

```python
# cv2 does not throw error allthough the path is invalid 
img = cv2.imread("../DATA/00-puppy.jpg")

print(type(img))  # if the image is not read successfully, then the type will be NoneType
```

```python
img.shape
```

## Display the image

```python
# show the image
# matplotlib expects in this order --> Red, Green, Blue (RGB)
# opencv expects in this order     --> Blue, Green, Red (BGR)

# convert the image from bgr to rgb
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
plt.show()
```

## Read a image in gray scale

```python
img_gray = cv2.imread("../DATA/00-puppy.jpg", cv2.IMREAD_GRAYSCALE)

img_gray.shape
```

```python
plt.imshow(img_gray, cmap="gray")
plt.show()
```

## Resize the image

```python
fix_img.shape
```

```python
new_img = cv2.resize(fix_img, (1000, 400))

plt.imshow(new_img)
plt.show()
```

## Resize in Width/Height Ratio

```python
w_ratio = 0.5
h_ratio = 0.5

new_img = cv2.resize(fix_img, (0, 0), fix_img, w_ratio, h_ratio)

plt.imshow(new_img)
plt.show()
```

## Flip Images

```python
new_img = cv2.flip(fix_img, 0)

plt.imshow(new_img)
plt.show()
```

```python
new_img = cv2.flip(fix_img, 1)

plt.imshow(new_img)
plt.show()
```

```python
new_img = cv2.flip(fix_img, -1)

plt.imshow(new_img)
plt.show()
```

## Save an Image

```python
type(fix_img)
```

```python
cv2.imwrite("totally_new.png", fix_img)
```

# Draw Shapes on Image

**Features:**
- Create a blank image
- Draw Rectangle/Circle/Line on the Image
- Fill the Rectangle/Circle
- Add text on the image
- Draw Polygon on the image

```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

## Create blank image

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
plt.show()
```

## Draw rectangle

```python
# remember: this is change the source image permanently
cv2.rectangle(blank_img, pt1=(384, 10), pt2=(500, 150), color=(0, 255, 0), thickness=10)

# show the image
plt.imshow(blank_img)
plt.show()
```

```python
cv2.rectangle(blank_img, pt1 = (200, 200), pt2 = (300, 300), color = (0, 0, 255), thickness = 10)

# show the image
plt.imshow(blank_img)
plt.show()
```

## Draw Circle

```python
cv2.circle(img=blank_img, center=(100, 100), radius=50, color=(255, 0, 0), thickness=8)

# show the image
plt.imshow(blank_img)
plt.show()
```

### Fill the circle

```python
cv2.circle(img=blank_img, center=(400, 400), radius=50, color=(184, 255, 51), thickness=-1)

# show the image
plt.imshow(blank_img)
plt.show()
```

## Draw line

```python
cv2.line(blank_img, pt1=(0, 0), pt2=(512, 512), color=(255, 51, 199), thickness=5)

# show the image
plt.imshow(blank_img)
plt.show()
```

## Add Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text="Hello", org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 255), thickness=3, lineType=cv2.LINE_AA)

# show the image
plt.imshow(blank_img)
plt.show()
```

## Custom Polygons

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int32)

plt.imshow(blank_img)
plt.show()
```

```python
vertices = np.array([ [100, 300], [200, 200], [400, 300], [200, 400] ], dtype=np.int32)

vertices.shape
```

```python
pts = vertices.reshape((-1, 1, 2))
pts.shape
```

```python
print(vertices)
print()
print(pts)
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255, 0, 0), thickness=5)

# show the image
plt.imshow(blank_img)
plt.show()
```

```python
import numpy as np

from statsmodels.graphics.tsaplots import plot_pacf, plot_acf

import matplotlib.pyplot as plt
```

## PACF

```python
# create noise by standard normal
x0 = np.random.randn(1_000)

# display that
plt.plot(x0)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

plot_pacf(x0, ax=ax);
ax.set_xlabel(r"lag ($\tau$)")
ax.set_ylim([-0.15, 1.15]);
```

```python
x1 = [0]

for i in range(1_000):
    x = 0.5 * x1[-1] + 0.1 * np.random.randn()
    x1.append(x)

x1 = np.array(x1)
```

```python
plt.plot(x1);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_pacf(x1, ax=ax)
ax.set_ylim([-0.09, 1.05]);

## From this plot, at lag = 1, we can get the desired outputs. Because of
## lag = 1, we can conclude that this is AR(1) process.
```

```python
x1 = [0]

for i in range(1_000):
    x = -0.5 * x1[-1] + 0.1 * np.random.randn()
    x1.append(x)

x1 = np.array(x1)
```

```python
plt.plot(x1);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_pacf(x1, ax=ax)
ax.set_ylim([-0.55, 1.05]);

## From this plot, at lag = 1, we can get the desired outputs. Because of
## lag = 1, we can conclude that this is AR(1) process.
```

```python
# generate the AR(2) process

x2 = [0, 0]
for i in range(1_000):
    x = 0.5 * x2[-1] - 0.3 * x2[-2] + 0.1 * np.random.randn()
    x2.append(x)

x2 = np.array(x2)
```

```python
plt.plot(x2);
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
plot_pacf(x2, ax=ax)
ax.set_ylim([-0.35, 1.1]);

# we get the result ar lag = 2, so, it is AR(2) process
```

```python
# next, generate a AR(5) process
x5 = [0 for i in range(5)]
for i in range(1_000):
    x = 0.5 * x5[-1] - 0.3 * x5[-2] - 0.6 * x5[-5] + 0.1 * np.random.randn() 
    x5.append(x)

x5 = np.array(x5)
```

```python
plt.plot(x5);
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
plot_pacf(x5, ax=ax)
ax.set_ylim([-0.7, 1.1]);

# we get the result ar lag = 5, so, it is AR(5) process
```

## ACF

```python
# iid noise
fig, ax = plt.subplots(figsize = (10, 5))

plot_acf(np.random.randn(1_000), ax=ax)
ax.set_ylim([-.15, 1.05]);
```

```python
errors = 0.1 * np.random.randn(1_000)
ma1 = []
for i in range(1_000):
    if i >= 1:
        x = 0.5 * errors[i-1] + errors[i]
    else:
        x = errors[i]
    ma1.append(x)

ma1 = np.array(ma1)
```

```python
plt.plot(ma1);
```

```python
fig, ax = plt.subplots(figsize = (10, 5))

plot_acf(ma1, ax=ax)
ax.set_ylim([-.15, 1.05]);

## MA(1) process from the graph
```

```python
errors = 0.1 * np.random.randn(1_000)
ma2 = []
for i in range(1_000):
    x = 0.5 * errors[i-1] - 0.3 * errors[i-2] + errors[i]
    ma2.append(x)

ma2 = np.array(ma2)
```

```python
plt.plot(ma2);
```

```python
fig, ax = plt.subplots(figsize = (10, 5))

plot_acf(ma2, ax=ax)
ax.set_ylim([-.22, 1.05]);

# from the plot, MA(2) process
```

```python
errors = 0.1 * np.random.randn(1_000)
ma3 = []
for i in range(1_000):
    x = 0.5 * errors[i-1] - 0.3 * errors[-2] + 0.7 * errors[i-3] + errors[i]
    ma3.append(x)

ma3 = np.array(ma3)
```

```python
plt.plot(ma3);
```

```python
fig, ax = plt.subplots(figsize = (10, 5))

plot_acf(ma3, ax=ax)
ax.set_ylim([-.15, 1.05]);

# from the plot, this is MA(3) process
```

```python
errors = 0.1 * np.random.randn(1_000)
ma6 = []
for i in range(1_000):
    x = 0.5 * errors[i-1] - 0.3 * errors[i-2] + 0.7 * errors[i-3] + \
          0.2 * errors[i-4] - 0.8 * errors[i-5] - 0.9 * errors[i-6] + errors[i]
    ma6.append(x)
ma6 = np.array(ma6)
```

```python
plt.plot(ma6);
```

```python
fig, ax = plt.subplots(figsize = (10, 5))

plot_acf(ma6, ax=ax)
ax.set_ylim([-.4, 1.05]);

# from the plot, this is MA(6) process
```

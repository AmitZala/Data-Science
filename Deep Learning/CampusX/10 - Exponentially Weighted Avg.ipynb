```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DailyDelhiClimateTrain.csv")[["date", "meantemp"]]
df.sample(5)
```

```python
plt.scatter(df["date"], df["meantemp"], color="red")
plt.show()
```

```python
# EWA calculation
x1 = df["meantemp"].ewm(alpha=0.1).mean()  # beta = 1 - alpha
```

```python
df["ewa"] = x1
```

```python
fig, axs = plt.subplots(1, 1, figsize = (14, 8))
plt.scatter(df["date"], df["meantemp"], color="red")
plt.plot(df["date"], x1, color="black", label="alpha=0.1")
plt.legend()
plt.xlabel("Dates")
plt.ylabel("Temp")
plt.xticks(rotation=90)
plt.show()
```
